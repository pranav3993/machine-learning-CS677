# K-Means Algorithm

Optimal Value to find K using Elbow Method:

```python
ssd = []
k = range(1,20)
for i in k:
    km = KMeans(n_clusters=i)
    km = km.fit(tdf_2)
    ssd.append(km.inertia_)
```

```python
plt.plot(k,ssd,'bx-')
plt.xlabel('Values of K') 
plt.ylabel('ssd') 
plt.title('Elbow Method')
plt.show()
```

```python
X = tdf_1[['value','WeekDay']]
x = pd.DataFrame(X)
x.shape
```

```python
km_n = KMeans(n_clusters=4).fit(x)
```

```python
km_n.labels_
```

```python
res = km_n.labels_
res = pd.DataFrame(res, columns = ['cluster'])
res.groupby('cluster').size()
```

# Outlier Detection using IQR:

```python
Q1 = np.percentile(tdf['value'], 25)
Q2 = np.percentile(tdf['value'], 50)
Q3 = np.percentile(tdf['value'], 75)
print(Q1)
print(Q2)
print(Q3)
```

```python
IQR = stats.iqr(tdf['value'])
IQR
```

```python
low_lim = Q1 - 1.5 * IQR  # LOWER LIMIT
upr_lim = Q3 + 1.5 * IQR  # UPPER LIMIT
print(low_lim)
print(upr_lim)
```

```python
outlier = []
for i in tdf['value']:
    if ((i>upr_lim or i<low_lim)):
        outlier.append(i)
print('These are the outliers we\'re looking for:', outlier)
print('The number of outliers present are:', len(outlier))
```

```python
tdf_3 = tdf_2.copy()
tdf_3
```

```python
i_frt =IsolationForest(n_estimators=50, max_samples='auto', contamination=float(0.1),max_features=1.0).fit(tdf_3[['value']])
```

```python
tdf_3.loc[tdf_3['outlier']==-1]              # All the 727 anomalies are listed below
```

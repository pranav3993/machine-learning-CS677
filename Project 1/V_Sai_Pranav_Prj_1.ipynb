```python
d_1['aapl_bs'] = np.where(d_1['Close'].iloc[:,:1].shift(-1) > d_1['Close'].iloc[:,:1], 1, -1)
```

```python
d_1['amzn_bs'] = np.where(d_1['Close'].iloc[:,1:2].shift(-1) > d_1['Close'].iloc[:,1:2], 1, -1)
```

```python
d_1['msft_bs'] = np.where(d_1['Close'].iloc[:,2:].shift(-1) > d_1['Close'].iloc[:,2:], 1, -1)
d_1
```

```python
x_lst = []
```

```python
x_lst = [[d_1['Close'].iloc[:,:1], d_1['High'].iloc[:,:1], d_1['Low'].iloc[:,:1], d_1['Open'].iloc[:,:1]]]
x = pd.DataFrame(x_lst)
x
```

```python
x_lst
```

```python
y = d_1[['aapl_bs']]
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2)
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)

knn_ft = knn.fit(x_train, y_train)

y_pred = knn.predict(x_test)
```

```python
metrics.accuracy_score(y_test, y_pred)
```
